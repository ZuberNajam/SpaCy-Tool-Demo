```python
# import the spacy library
import spacy
```

```python
# create the nlp object with the English language class
nlp = spacy.load("en_core_web_sm")
```

```python
from IPython.display import Image
Image("pipeline.png")
```

```python
# check which components are in the processing pipeline
nlp.pipe_names
```

```python
# process a string of text with the nlp object in order to
# have spaCy create a doc object
doc=nlp("Ali and Khalid went to Dubai for spring break.")
```

Demonstration Part 1: Tokenizer - Lemmatizer - Part-of-Speech Tagger

```python
# a doc object is like a normal python sequence which 
# can be iterated over
print("Token ", " Lemma", " Part-of-speech")
for token in doc:
    print(token.text,"|", token.lemma_, "|", token.pos_)
```

```python
# for more information, we can use spacy's explain attribute
for token in doc:
    print(spacy.explain(token.pos_))
```

```python
from spacy import displacy
displacy.render(doc, style='dep', jupyter=True)
```

Demonstration Part 2: Entity Recognizer

```python
for ent in doc.ents:
    print(ent.text)
```

```python
displacy.render(doc, style='ent')
```

Demonstration Part 3: Adding custom components to the pipeline

```python
from spacy.language import Language

@Language.component("token_count")

# define the custom component be creating a python function
def token_count(doc):
    print("The number of tokens present:", len(doc))
    return doc 
```

```python
nlp.add_pipe("token_count")
```

```python
print("Pipeline components:", nlp.pipe_names)
```
